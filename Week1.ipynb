```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import cv2 as cv
from google.colab.patches import cv2_imshow # Use For Image display
from skimage import io
from skimage import transform
from PIL import Image
```

```python
#Creating list to store url from several images
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg","https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg","https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg"]

#Read and Display Image
for url in urls:
  image = io.imread(url)                          #Read Image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #Resize Image to Half Size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)  #Convert Image to RGB
  final_frame = cv.hconcat((image, image_2))      #Concatenate Image
  cv2_imshow(final_frame)                         #Display Image
  print("\n")
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ", tinggi, "x", lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#Creating Horizontal Line in the middle of the image
for y in range (lebar) :
  image_3[int((tinggi)/2), y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#Creating Vertical Line in the middle of the image
for y in range (tinggi) :
  image_3[y, int ((lebar)/2)] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#Creating Vertical Line in the middle of the image
min_side = min(tinggi, lebar)
for y in range (min_side) :
  image_3[y, y] = [255,255,255]

for y in range (min_side) :
  image_3[y, lebar-1-y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
y_center = int(tinggi/2)
line_length = 100
x_start = int((lebar - line_length)/2)
x_end = x_start + line_length
for x in range(x_start, x_end):
    image_3[y_center, x] = [255, 255, 255]

final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

box_size = 100
x_center = lebar// 2
y_center = tinggi // 2
x1 = x_center - box_size // 2
y1 = y_center - box_size // 2
x2 = x_center + box_size // 2
y2 = y_center + box_size // 2

# kotak
for y in range(y1, y2):
    for x in range(x1, x2):
        image_3[y, x] = [255, 255, 255]

final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```
