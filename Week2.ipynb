```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
img = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/kuceng.jpg')
plt.imshow(img)
```

```python
img2 = cv.resize(cv.cvtColor(img, cv.COLOR_BGR2RGB), (512,1024))
plt.imshow(img2)
```

```python
img2 = cv.flip(cv.cvtColor(img, cv.COLOR_BGR2RGB), 0)

#Showing plot with bigger canvas
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.imshow(img2)
```

```python
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_img)
```

```python
cv.rectangle(black_img, pt1=(384,0), pt2=(510,150), color=(0,255,0), thickness=10)
plt.imshow(black_img)
```

```python
cv.rectangle(black_img, pt1=(200,200), pt2=(300,300), color=(0,0,255), thickness=15)
plt.imshow(black_img)
```

```python
cv.circle(black_img, center=(100,100), radius=50, color=(255,0,0), thickness=8)
plt.imshow(black_img)
```

```python
cv.line(black_img, pt1=(0,0), pt2=(511,511), color=(255,255,255), thickness=5)
plt.imshow(black_img)
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='PCVK', org=(10,500), fontFace=font, fontScale=4, color=(255,255,0), thickness=2, lineType=cv.LINE_AA)
plt.imshow(black_img)
```

```python
black_img2 = np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_img2)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]], dtype=np.int32)
vertices
```

```python
pts = vertices.reshape((-1,1,2))
pts
```

```python
cv.polylines(black_img2, [pts], isClosed=True, color=(255,0,0), thickness=5)
plt.imshow(black_img2)
```
