```python
import random
img_path = '/content/drive/MyDrive/PCVK_MM/Images/kuceng.jpg'
img = cv.imread(img_path)
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
plt.figure(figsize=(15, 8))
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
# Create images with only red and green channels, and only green and blue channels
rg_img = img_rgb.copy()
rg_img[:, :, 2] = 0 # Set blue channel to 0

gb_img = img_rgb.copy()
gb_img[:, :, 0] = 0 # Set red channel to 0

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(rg_img)
axs[0].axis('off')

axs[1].imshow(gb_img)
axs[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
cropped = img_rgb[10:100, 10:100]
plt.imshow(cropped)
plt.show()
```

```python
green_channel = img_rgb[5:30, :, 1]
plt.imshow(green_channel, cmap='Greens_r')
plt.title('Rows 5-30, All columns (Green only)')
plt.show()
```

```python
img_boxes = img_rgb.copy()
for _ in range(5):
    x1 = random.randint(0, img.shape[1] - 100)
    y1 = random.randint(0, img.shape[0] - 100)
    x2 = x1 + random.randint(20, 100)
    y2 = y1 + random.randint(20, 100)
    color = tuple(random.randint(0, 255) for _ in range(3))
    cv.rectangle(img_boxes, (x1, y1), (x2, y2), color, 2)

plt.imshow(img_boxes)
plt.axis('off')
plt.show()
```

```python
import os

img_path = '/content/drive/MyDrive/PCVK_MM/Images/seno.jpeg'

if not os.path.exists(img_path):
    print(f"Error: File not found at {img_path}")
else:
    img = cv.imread(img_path)
    if img is None:
        print(f"Error: Could not read image from {img_path}")
    else:
        img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
        plt.imshow(img_rgb)
        plt.show()
```

```python
img_face = img_rgb.copy()

cv.rectangle(img_face, (1300, 200), (2000, 950), (0, 255, 0), 2)  # Wajah
cv.circle(img_face, (1540, 540), 60, (255, 0, 0), 5)  # Mata kiri
cv.circle(img_face, (1750, 545), 60, (0, 255, 0), 5)  # Mata kanan
cv.circle(img_face, (1645, 765), 70, (0, 0, 255), 5)  # Mulut

plt.imshow(img_face)
plt.show()
```

```python
img_blue = img.copy()
cv.rectangle(img_blue, (0, img.shape[0]-100), (100, img.shape[0]-1), (255, 0, 0), -1)

img_blue_rgb = cv.cvtColor(img_blue, cv.COLOR_BGR2RGB)
plt.imshow(img_blue_rgb)
plt.show()
```

```python
img_text = img_blue.copy()
cv.putText(img_text, "Seno.jpg", (80, 90), cv.FONT_HERSHEY_SIMPLEX, 3, (0, 0, 0), 3)
img_text_rgb = cv.cvtColor(img_text, cv.COLOR_BGR2RGB)

plt.imshow(img_text_rgb)
plt.axis('off')
plt.show()
```
